```python
import pandas as pd
import numpy as np
df =pd.read_csv("nba.csv")
```

```python
df.head()
```

```python
#The numpy where method
df['Position_PG'] = np.where(df['Position'] == "PG", 1, 0)
df.head()
```

```python
#Applying a function
def my_recode(Position):
    if Position == "PG":
        return 1
    else:
        return 0
df['Position_PG'] = df['Position'].apply(my_recode)
df.head()
```

```python
#apply lambda function
df['Position_PG'] = df['Position'].apply(lambda x: 1 if x == "PG" else 0)
df.head()
```

```python
df.Position.unique()
```

```python
#replace values from a list
position=['PG', 'SF', 'SG', 'PF', 'C', np.nan]
position_num= [1,2,3,4,5,99]

df['Position_Num'] = df['Position'].replace(position, position_num)
df.head()
```

```python
#use map function
look_tb={'PG':1, 'SF':2, 'SG':3, 'PF':4, 'C':5, np.nan:99}
df['Position_Num'] = df['Position'].map(look_tb)
df.head()
```
